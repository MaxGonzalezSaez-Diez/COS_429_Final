```
import json
# Opening JSON file
# results_json = open('/content/AlphaPose/examples/res/alphapose-results.json')
results_json = open('/content/drive/Shareddrives/CV_videos/Video Results/video.json')
data = json.load(results_json)

old_f = data[0]["image_id"]
count = 1
for i in range(1, len(data)):
    new_f = data[i]["image_id"]
    if new_f != old_f:
        print("filename: " + str(old_f) + " -- #ppl: " + str(count))
        count = 1
        old_f = new_f
    else:
        count+=1

print("filename: " + str(old_f) + " -- #ppl: " + str(count))
```

```
import os

# mount drive
from google.colab import drive
drive.mount("/content/drive")

processed_dir = '/content/drive/Shareddrives/CV_videos/Video_Processed'
results_dir = '/content/drive/Shareddrives/CV_videos/Video_Results'

processed_files = []
results_files = []

# Iterate through the subdirectories in the processed directory
for subdir, _, files in os.walk(processed_dir):
    # Iterate through the files in each subdirectory
    for file in files:
        # If the file ends with .mp4, add its basename to the processed_files list
        if file.endswith('.mp4'):
            processed_files.append(os.path.splitext(file)[0])

# Iterate through the subdirectories in the results directory
for subdir, _, files in os.walk(results_dir):
    # Iterate through the files in each subdirectory
    for file in files:
        # If the file ends with .json, add its basename (with .json stripped) to the results_files list
        if file.endswith('.json'):
            results_files.append(os.path.splitext(file)[0])

# Check which processed files have a matching basename in the results files
non_matching_files = list(set(processed_files) - set(results_files))
matching_files = list(set(processed_files) & set(results_files))

# Print the matching files
print(len(matching_files), len(non_matching_files))

```

# Move back those that have been moved but not generated

```
import shutil
path_to_video_results = "/content/drive/Shareddrives/CV_videos/Video_Results/"
path_to_sports = "/content/drive/Shareddrives/CV_videos/Video_Processed/"
path_to_video_trimmed = "/content/drive/Shareddrives/CV_videos/Video_Trimmed/"

for folder_sport in os.listdir(path_to_sports):
    print("Doing " + str(folder_sport) + "...")
    for video_processed in os.listdir(os.path.join(path_to_sports, folder_sport)):
        json_name = video_processed.strip(".mp4") + str(".json")
        if os.path.isfile(os.path.join(path_to_video_results, folder_sport, json_name)):
            print(json_name + " correctly processed.")
            pass
        else:
            print("Moving this file back: ", json_name)
            path_source = os.path.join(path_to_sports, folder_sport, video_processed)
            path_destination = os.path.join(path_to_video_trimmed, folder_sport, video_processed)
            shutil.move(path_source, path_destination)
```
